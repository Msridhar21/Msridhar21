# An exploratory analysis of crimes against women in india from 2001 to 2014 in different states

### This project on crime on women in india from 2001 to 2014 to create awareness among the people in society about the situation of women

## 1.data preparation and cleaning

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
sns.set()
df=pd.read_csv('crimes_against_women_2001-2014.csv')
#df.describe()
df
```

```python
df.shape
```

```python
print(df.isnull().sum())
```

```python
df.drop(['Unnamed: 0'],axis=1,inplace=True)
df
```

```python
dist_count=len(df['DISTRICT'].unique())
dist_count
```

```python
df.drop(['DISTRICT'],axis=1,inplace=True)
df
```

```python
df=df.rename(columns={
    'Kidnapping and Abduction':'Kidnapping_and_Abduction',
    'Dowry Deaths':'Dowry_Deaths',
    'Assault on women with intent to outrage her modesty':'Assault_on_women_with_intent_to_outrage_her_modesty',
    'Insult to modesty of Women':'Insult_to_modesty_of_Women',
    'Cruelty by Husband or his Relatives':'Cruelty_by_Husband_or_his_Relatives',
    'Importation of Girls':'Importation_of_Girls'
                     }
            )
df
```

```python
print(df['STATE/UT'].unique())
print(len(df['STATE/UT'].unique()))
```

```python
def cap_state(word):
    word=word['STATE/UT'].strip()
    word=word.upper()
    return word
df['STATE/UT']=df.apply(cap_state,axis=1)
df['STATE/UT']=df['STATE/UT'].replace(
    {
        'A&N ISLANDS':'A & N ISLANDS',
        'D&N HAVELI':'D & N HAVELI',
        'DELHI UT':'DELHI'
    }
                      )
    
print(df['STATE/UT'].unique())
print(len(df['STATE/UT'].unique()))
df
```

## 2.analysis and visualization of the data

```python
norape=df.Rape.sum()
nokidnapped=df.Kidnapping_and_Abduction.sum()
nodowry=df.Dowry_Deaths.sum()
noassault=df.Assault_on_women_with_intent_to_outrage_her_modesty.sum()
noinsulted=df.Insult_to_modesty_of_Women.sum()
nocruelty=df.Cruelty_by_Husband_or_his_Relatives.sum()
noimported=df.Importation_of_Girls.sum()
total_no=0
total_no=norape+nokidnapped+nodowry+noassault+nocruelty+noinsulted+noimported
total_no
```

### It is clear that nearly 5 million women were suffered from various crimes against them from 2001 to 2014

### now lets analyse with some statistics 

```python
fig, axes = plt.subplots(2, 3, figsize=(25, 12))

axes[0,0].bar(df.Year,df.Rape,color='r')
axes[0,0].set_title('Cases of Rape in India from 2001 to 2014')

axes[0,1].bar(df.Year,df.Kidnapping_and_Abduction,color='black')
axes[0,1].set_title('Cases on Kidnapping and abduction of women in India from 2001 to 2014')

axes[0,2].bar(df.Year,df.Dowry_Deaths,color='blue')
axes[0,2].set_title('Cases on Dowry deaths of women in India from 2001 to 2014')

axes[1,0].bar(df.Year,df.Insult_to_modesty_of_Women,color='orange')
axes[1,0].set_title('Cases of Insult on women in India from 2001 to 2014')

axes[1,1].bar(df.Year,df.Cruelty_by_Husband_or_his_Relatives,color='gray')
axes[1,1].set_title('Cases on Domestic voilence in India from 2001 to 2014')

axes[1,2].bar(df.Year,df.Importation_of_Girls,color='pink')
axes[1,2].set_title('Cases on Importation of girls in India from 2001 to 2014')
```

### from the graphs 2 things are  clear 
### 1.Cases are increasing every year except the importation cases
### 2.2014 is the year with maximum crimes committed against women
